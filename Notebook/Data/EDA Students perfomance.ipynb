### 3.8 Adding columns for "Total Score" and "Average"

```python
#Creating a new columns to get more insights from the dataset
df['total_score'] = df['math_score'] + df['reading_score'] + df['writing_score']

#Create a column for averge scores of the students
df['average_score'] = df['total_Score'] / 3

df[['total_score','average_score']]
```

```python
# Finding how many students have got full marks(100)

math_full = df[df['math_score'] == 100].shape[0]
reading_full = df[df['reading_score'] == 100].shape[0]
writing_full = df[df['writing_score'] == 100].shape[0]

print(f'Number of students with full marks in Maths: {math_full}')
print(f'Number of students with full marks in Reading: {reading_full}')
print(f'Number of students with full marks in Writing: {writing_full}')
```

```python
# Finding how many students have got below 35
math_35 = df[df['math_score'] < 35].shape[0]
reading_35 = df[df['reading_score'] < 35].shape[0]
writing_35 = df[df['writing_score'] < 35].shape[0]

print(f'Number of students with less than 20 marks in Maths: {math_35}')
print(f'Number of students with less than 20 marks in Reading: {reading_35}')
print(f'Number of students with less than 20 marks in Writing: {writing_35}')
```

#####  Insights
 - From above values we get students have performed the worst in Maths 
 - Best performance is in reading section

#### Insights 
- Group E students have scored the highest marks. 
- Group A students have scored the lowest marks. 
- Students from a lower Socioeconomic status have a lower avg in all course subjects
